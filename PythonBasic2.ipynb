```python
l=[1,3,4,6,7,8,9,2,3,4,5]
sumod=0
sumev=0
for i in l:
    if(i%2!=0):
        sumod += i
    else:
        sumev += i
print(sumod)
print(sumev)
```

# For loop with else

```python
l=[2,6,8,5,7,4,9,8,4,7,3,5]
odd_sum=0
even_sum=0
for i in l:
    if(i%2==0):
        even_sum += i
        if even_sum>20:
            break
    else:
        odd_sum += i
        if(odd_sum>20):
            break
else:
    print("Sum is smaller then 20")
print("done")
```

```python
#For Else
l=[1,2,3,4,5,6]
sum1=0
for i in l:
    sum1 += i
else:
    print(sum1)
```

# While Loop

```python
# Sum of first 10 natural numbers
n=10
sum1=0
i=1
while i<=n:
    sum1 += i
    i += 1
print(sum1)
```
