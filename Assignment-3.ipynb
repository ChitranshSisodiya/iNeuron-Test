1.1 Write a Python Program to implement your own myreduce() function which works exactly
like Python's built-in function reduce()

```python
def reduce_fun(fun,args):
    if(type(args)==int):
        return fun(args)
    elif(type(args)==list):
        sum1=0
        for i in args:
            sum1 += fun(i)
        return sum1
def fun(x):
    return x**2
lst=[1,2,3,4,5]
a=5
print(reduce_fun(fun,a))
```

```python
print(reduce_fun(fun,lst))
```

1.2 Write a Python program to implement your own myfilter() function which works exactly
like Python's built-in function filter()

```python
# odd numbers to display
def filter_fun(fun2,lst):
    lst2=[]
    for i in lst:
        if(fun2(i)):
            lst2.append(i)
    return lst2
fun2=lambda x : True if x%2!=0 else False
```

```python
lst=[1,2,3,4,5,6,7,8,9]
print(filter_fun(fun2,lst))
```
