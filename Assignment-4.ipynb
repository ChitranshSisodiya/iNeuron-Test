1.1 Write a Python Program(with class concepts) to find the area of the triangle using the below formula.

area = (s*(s-a)*(s-b)*(s-c)) ** 0.5

Function to take the length of the sides of triangle from user should be defined in the parent class and function to calculate the area should be defined in subclass.

```python
class inp:
    def __init__(self):
        self.side1=float(input("Enter Length of First Side :- "))
        self.side2=float(input("Enter Length of Second Side :- "))
        self.side3=float(input("Enter Length of Third Side :- "))
class out(inp):
    def __init__(self):
        super().__init__()
    def output(self):
        s=float((self.side1+self.side2+self.side3)/2)
        area = (s*(s-self.side1)*(s-self.side2)*(s-self.side3))**0.5
        print("The Area Of The Triangle is :  {}".format(area))
a=out()
a.output()
```

1.2 Write a function filter_long_words() that takes a list of words and an integer n and returns the list of words that are longer than n.

```python
def filter_long_words(lst,n):
    for i in lst:
        if(len(i)>n):
            print(i)
lst=['fggahgshsgasfgsgs','fgwffafawfafffs','dfsf','sdfafsda','gfdasg','sad','gffsdagf','dfsfdsfsfsdfsdf']
n=int(input("Enter The Minimun Length Of String : "))
filter_long_words(lst,n)
```

2.1 Write a Python program using function concept that maps list of words into a list of integers representing the lengths of the corresponding words.

Hint: If a list [ ab,cde,erty] is passed on to the python function output should come as [2,3,4]

Here 2,3 and 4 are the lengths of the words in the list.

```python
def maplist(words):
    lists=list(map(lambda x : len(x),words))
    return lists
words=['fggahgshsgasfgsgs','fgwffafawfafffs','dfsf','sdfafsda','gfdasg','sad','gffsdagf','dfsfdsfsfsdfsdf']
print("The Map List is : {}".format(maplist(words)))
```
