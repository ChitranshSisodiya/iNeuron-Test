1. Create the below pattern using nested for loop in Python.

```python
def pattern(n):
    count=1
    for i in range(0,n):
        for j in range(0,n):
            if(count>j):
                print("*",end=' ')
        print()
        count +=1
    count=n-1
    for i in range(0,n):
        for j in range(0,n):
            if(count>j):
                print('*',end=' ')
        print()
        count -= 1            
pattern(5)
```

2. Write a Python program to reverse a word after accepting the input from the user.

```python
word=str(input("Enter A Word : "))
for i in range(-1,-len(word)-1,-1):
    print(word[i],end='')
```
